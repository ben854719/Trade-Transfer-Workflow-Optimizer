```python
from IPython import display
import os, requests, nltk, cv2, io
from urllib.parse import quote
from PIL import Image
from langgraph.graph import StateGraph, END
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_core.messages import HumanMessage
from unstructured.partition.html import partition_html
from unstructured.partition.auto import partition
from google.colab import userdata
from google.colab.patches import cv2_imshow
from IPython.display import Markdown
from mcp.server.fastmcp import FastMCP

# Load avatar image
avatar_path = "Made with insMind-Businessman.jpg"

# Download NLTK data
nltk.download('punkt', quiet=True)
nltk.download('averaged_perceptron_tagger', quiet=True)

# Load Gemini API key
api_key = userdata.get("Ben856")
if not api_key:
    raise ValueError("Ben856 secret not found. Please set your API key in Colab Secrets.")

# Initialize Gemini model
gemini_model = ChatGoogleGenerativeAI(model='gemini-2.5-flash', api_key=api_key)

# Node 1: Fetch and process content
def fetch_and_process_content(state):
    url = state.get("url")
    image_url = state.get("image_url")

    if url:
        try:
            safe_url = quote(url, safe=":/?=&")
            response = requests.get(safe_url, timeout=20)
            response.raise_for_status()
            elements = partition_html(text=response.text)
            text_content = "\n".join([str(el) for el in elements])
            return {"url": url, "text": text_content, "content_type": "text"}
        except Exception as e:
            return {"url": url, "text": f"Error fetching URL: {e}", "content_type": "error"}

    elif image_url:
        try:
            response = requests.get(image_url, stream=True)
            response.raise_for_status()
            img = Image.open(io.BytesIO(response.content))
            if img.mode != 'RGB':
                img = img.convert('RGB')
            img_path = "temp_image.jpg"
            img.save(img_path)
            elements = partition(filename=img_path)
            text_content = "\n".join([str(el) for el in elements])
            os.remove(img_path)
            return {"image_url": image_url, "text": text_content, "content_type": "image"}
        except Exception as e:
            return {"image_url": image_url, "text": f"Error processing image: {e}", "content_type": "error"}

    else:
        return {"text": "Error: No URL or image URL provided.", "content_type": "error"}

# Node 2: Analyze content with Gemini
def analyze_content(state):
    url = state.get("url")
    image_url = state.get("image_url")
    text = state.get("text", "No content found.")
    content_type = state.get("content_type", "unknown")
    source = url if content_type == "text" else image_url if content_type == "image" else "unknown source"

    prompt_text = f"""
You're an expert analyst reviewing content from: {source}

Content Type: {content_type}

Content:
{text}

Please provide:
- A concise summary
- Key topics or themes
- Any notable insights or implications
"""

    try:
        response = gemini_model.invoke([HumanMessage(content=prompt_text)])
        analysis_result = response.content
    except Exception as e:
        analysis_result = f"Error during Gemini analysis: {e}"

    return {"url": url, "image_url": image_url, "text": text, "analysis": analysis_result, "content_type": content_type}

# Node 3: Avatar commentary
def avatar_commentary(state):
    analysis = state.get("analysis", "No analysis available.")
    image_url = state.get("image_url")
    url = state.get("url")
    content_type = state.get("content_type", "unknown")
    source = image_url if content_type == "image" else url if content_type == "text" else "unknown source"

    prompt = f"""
You're a confident, expressive avatar speaking directly from the content below.

Source: {source}
Content Type: {content_type}

Here's what you've observed:
{analysis}

Now speak as if you're coming alive from the image or page. Respond with:
- A short voice line (1–2 sentences) as if you're addressing the viewer
- A longer commentary (2–4 sentences) that reflects your personality and insights
- Optional emotional cue (e.g., confident, curious, amused)
"""

    try:
        response = gemini_model.invoke([HumanMessage(content=prompt)])
        avatar_output = response.content
    except Exception as e:
        avatar_output = f"Error generating avatar commentary: {e}"

    return {**state, "avatar_commentary": avatar_output}

# LangGraph workflow
workflow = StateGraph(dict)
workflow.add_node("fetch_and_process", fetch_and_process_content)
workflow.add_node("analyze", analyze_content)
workflow.add_node("avatar", avatar_commentary)

def route_content(state):
    content_type = state.get("content_type")
    if content_type in ["text", "image"]:
        return "analyze"
    else:
        return "end"

workflow.set_entry_point("fetch_and_process")
workflow.add_conditional_edges("fetch_and_process", route_content, {"analyze": "analyze", "end": END})
workflow.add_edge("analyze", "avatar")
workflow.add_edge("avatar", END)

executable_graph = workflow.compile()

# Display avatar image and commentary (no voice)
def display_avatar_output(state):
    commentary = state.get("avatar_commentary", "No avatar output.")
    source = state.get("url") or state.get("image_url") or "Unknown source"

    # Show avatar image
    try:
        avatar_img = cv2.imread(avatar_path)
        if avatar_img is not None:
            print("🧍 Avatar:")
            cv2_imshow(avatar_img)
        else:
            print(f"⚠️ Could not load avatar image from {avatar_path}")
    except Exception as e:
        print(f"Error displaying avatar image: {e}")

    # Show commentary
    display.display(Markdown(f"### 🗣️ Avatar Speaks from {source}\n\n{commentary}"))

# Inputs to process (articles)
inputs = [
    {"url": "https://simplywall.st/stocks/ca/banks/tsx-bmo/bank-of-montreal-shares"},
    {"url": "https://simplywall.st/stocks/ca/banks/tsx-cm/canadian-imperial-bank-of-commerce-shares/news/is-cibcs-recent-25-rally-justified-after-strong-quarterly-ea"},
    {"url": "https://simplywall.st/stocks/ca/banks/tsx-td/toronto-dominion-bank-shares/news/how-investors-are-reacting-to-toronto-dominion-bank-tsxtd-un"},
    {"url": "https://simplywall.st/stocks/ca/banks/tsx-bns/bank-of-nova-scotia-shares/news/scotiabank-tsxbns-valuation-in-focus-after-new-bond-offering"},
    {"url": "https://simplywall.st/stocks/ca/banks/tsx-na/national-bank-of-canada-shares/news/a-fresh-look-at-national-bank-of-canada-tsxna-valuation-foll"},
]

# Run workflow
for input_data in inputs:
    if "url" in input_data:
        print(f"\n Processing URL: {input_data['url']}")
    elif "image_url" in input_data:
        print(f"\n Processing Image URL: {input_data['image_url']}")
    else:
        print("\n Invalid input data.")
        continue

    final_state = executable_graph.invoke(input_data)

    if final_state.get("content_type") in ["text", "image"]:
        display_avatar_output(final_state)
    else:
        print(f"\n Processing failed: {final_state.get('text', 'Unknown error.')}")
```
