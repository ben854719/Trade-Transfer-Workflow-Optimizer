```python
import pandas as pd
```

```python
# Import the dataset for the client stock portfolio.
INDEX_CA_XTSE_GSPTSE = pd.read_csv('/content/INDEX_CA_XTSE_GSPTSE.csv')
display(INDEX_CA_XTSE_GSPTSE.head(22))
```

```python
# Remove any duplication in the dataset.
INDEX_CA_XTSE_GSPTSE = INDEX_CA_XTSE_GSPTSE.drop_duplicates()
display(INDEX_CA_XTSE_GSPTSE.head(22))
```

```python
# Rename the column for the dataset.
INDEX_CA_XTSE_GSPTSE = INDEX_CA_XTSE_GSPTSE.rename(columns={'Date': 'Month_Date', 'Open': 'Opening_Stock_Price', 'High': 'Hight_Price_Stock', 'Low': 'Low_Stock_Price', 'Close': 'Close_Stock_Price'})
display(INDEX_CA_XTSE_GSPTSE.head(22))
```

```python
# Calculate the total price for each row (sum of Open, High, Low, Close)
INDEX_CA_XTSE_GSPTSE['Total_Row_Price'] = INDEX_CA_XTSE_GSPTSE['Opening_Stock_Price'].astype(str).str.replace(',', '').astype(float) + \
                                          INDEX_CA_XTSE_GSPTSE['Hight_Price_Stock'].astype(str).str.replace(',', '').astype(float) + \
                                          INDEX_CA_XTSE_GSPTSE['Low_Stock_Price'].astype(str).str.replace(',', '').astype(float) + \
                                          INDEX_CA_XTSE_GSPTSE['Close_Stock_Price'].astype(str).str.replace(',', '').astype(float)

# Calculate the total price for each column (sum of all values in each price column)
total_column_prices = {
    'Total_Opening_Stock_Price': INDEX_CA_XTSE_GSPTSE['Opening_Stock_Price'].astype(str).str.replace(',', '').astype(float).sum(),
    'Total_Hight_Price_Stock': INDEX_CA_XTSE_GSPTSE['Hight_Price_Stock'].astype(str).str.replace(',', '').astype(float).sum(),
    'Total_Low_Stock_Price': INDEX_CA_XTSE_GSPTSE['Low_Stock_Price'].astype(str).str.replace(',', '').astype(float).sum(),
    'Total_Close_Stock_Price': INDEX_CA_XTSE_GSPTSE['Close_Stock_Price'].astype(str).str.replace(',', '').astype(float).sum()
}

display(INDEX_CA_XTSE_GSPTSE.head())
print("\nTotal Price for Each Column:")
for col, total in total_column_prices.items():
    print(f"{col}: {total:,.2f}")
```
